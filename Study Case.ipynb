Menghitung Upah Pegawai

Pak Yanto adalah manajer keuangan di sebuah coffe shop. Dia ingin menghitung upah karyawan di coffeshopnya.
Daftar karyawan Pak Yanto adalah sebagai berikut:

**1A.**
Hitunglah gaji kotor masing-masing karyawan dengan ketentuan sebagai berikut:
1. Upah per jam untuk karyawan adalah 75000.
2. Jika pengalaman kerja 1-2 tahun maka upah per jam ada kenaikan sebesar 5%, jika pengalaman kerja lebih dari 2 tahun maka upah per jam ada kenaiakn sebesar 10%.

Gaji Kotor = Total jam kerja x Upah per jam

**Contoh Output:**

Gaji (Kotor) Sholeh adalah  Rp 10000000


```python
# Jawab disini

# definisikan upah per jam sebagai variabel konstant
UPAH_PERJAM = 75000

class Karyawan :
  def __init__ (self, nama, pengalaman, jam_kerja, jam_lembur, gaji_kotor = 0, upah_lembur =0, total_gaji =0, pajak =0, potongan_bpjs = 0, netto =0) :
    self.nama = nama
    self.pengalaman = pengalaman
    self.jam_kerja = jam_kerja
    self.jam_lembur = jam_lembur
    self.gaji_kotor = gaji_kotor
    self.upah_lembur = upah_lembur
    self.total_gaji = total_gaji
    self.pajak = pajak
    self.potongan_bpjs = potongan_bpjs
    self.netto = netto

  def gajiKotor (selft, upah_perjam, pengalaman, jam_kerja) :
    gaji_kotor = 0

    if (pengalaman <= 2):
      #Jika pengalaman 1- 2 tahun upah perjam naik 5%
      upah_perjam_kenaikan = (105/100) * upah_perjam
      gaji_kotor = jam_kerja * upah_perjam_kenaikan

    else :
      #Jika pengalaman lebih 2 tahun upah perjam naik 10%
      upah_perjam_kenaikan = (110/100) * upah_perjam
      gaji_kotor = jam_kerja * upah_perjam_kenaikan

    return gaji_kotor

  def tampilGajiKotor (selft, nama, hasil) :
    print(f"Gaji (Kotor) {nama} adalah  Rp {hasil}")

  def updahLembur (selft, jam_lembur, lembur_perjam ) :
    return jam_lembur * lembur_perjam

  def tampilUpahLembur (selft, nama, hasil) :
    print(f"Gaji Lembur {nama} adalah  Rp {hasil}")

  def totalGaji (selft, gaji_kotor, upah_lembur) :
    return gaji_kotor + upah_lembur

  def tampilTotalGaji (selft, nama, hasil) :
    print(f"Gaji Kotor+Lembur {nama} adalah  Rp {hasil}")

  def hitungPajak (selft, total_gaji) :
    return total_gaji * 0.05

  def tampilPajak (selft, nama, hasil) :
    print(f"Total Pajak {nama} adalah  Rp {hasil}")

  def hitungBPJS (selft, total_gaji) :
    return total_gaji * 0.03

  def tampilBPJS (selft, nama, hasil) :
    print(f"Potongan BPJS {nama} adalah  Rp {hasil}")

  def hitungNetto (selft, total_gaji, pajak, potongan_bpjs) :
    # Total Upah (Netto) = Total Upah - (Pajak + BPJS)

    return total_gaji - (pajak + potongan_bpjs)

  def tampilNetto (selft, nama, hasil) :
    print(f"Total Upah (Netto) {nama} adalah  Rp {hasil}")



# Gaji Kotor Sholeh
# definisikan nama variabel untuk menampung nama karyawan 1 atau Sholeh
karyawan1 = Karyawan('Sholeh', 3, 160, 10) #Variabel ditampung di objek karyawan1

# definisikan variabel untuk menampung jam kerja karyawan 1
#Variabel ditampung di objek karyawan1

# hitung Gaji Kotor karyawan 1
karyawan1.gaji_kotor = karyawan1.gajiKotor(UPAH_PERJAM, karyawan1.pengalaman, karyawan1.jam_kerja)

# tampilkan Gaji Kotor karyawan 1
karyawan1.tampilGajiKotor(karyawan1.nama, karyawan1.gaji_kotor)

# untuk karyawan yang lain, lakukan seperti code di atas
#....

# Gaji Kotor Galih
karyawan2 = Karyawan('Galih', 2, 144, 20)
karyawan2.gaji_kotor = karyawan2.gajiKotor(UPAH_PERJAM, karyawan2.pengalaman, karyawan2.jam_kerja)
karyawan2.tampilGajiKotor(karyawan2.nama, karyawan2.gaji_kotor)

# Gaji Kotor Dewi
karyawan3 = Karyawan('Dewi', 1, 160, 10)
karyawan3.gaji_kotor = karyawan3.gajiKotor(UPAH_PERJAM, karyawan3.pengalaman, karyawan3.jam_kerja)
karyawan3.tampilGajiKotor(karyawan3.nama, karyawan3.gaji_kotor)

# Gaji Kotor Siska
karyawan4 = Karyawan('Siska', 4, 120, 30)
karyawan4.gaji_kotor = karyawan4.gajiKotor(UPAH_PERJAM, karyawan4.pengalaman, karyawan4.jam_kerja)
karyawan4.tampilGajiKotor(karyawan4.nama, karyawan4.gaji_kotor)


# Gaji Kotor Rizki
karyawan5 = Karyawan('Rizky', 1, 144, 20)
karyawan5.gaji_kotor = karyawan5.gajiKotor(UPAH_PERJAM, karyawan5.pengalaman, karyawan5.jam_kerja)
karyawan5.tampilGajiKotor(karyawan5.nama, karyawan5.gaji_kotor)
```

1B. Hitunglah Gaji Lembur dari seluruh karyawan

1. Upah lembur per jam = 50000
2. Upah Lembur = Jam Lembur x Upah lembur per jam
